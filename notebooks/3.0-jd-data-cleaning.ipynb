# New DataFrame

- First we will generate a new column with a new metric, **Student spending**. This will assess how much of the revenue per student is going to classroom instruction as opposed to miscellaneous costs.

***Student Spending = Instructional Expenditure per FTE / Net Tuition Revenue per FTE***
- Above 1.0 would indicate the university spends more than the revenue they receive per student, below 1 would indicate that the university spends less than it gains in revenue per student.
- Then, we will create a new dataframe with the headers renamed so we can display the data on our webapp.

```python
df["STUSPEND"] = df["INEXPFTE"] / df["TUITFTE"]
```
