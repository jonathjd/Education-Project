```python
# import necessary libraries
import pandas as pd
```

```python
# read in data to pandas dataframe
df = pd.read_csv("/Most-Recent-Cohorts-All-Data-Elements.csv")
df.head()
```

# Filtering out columns

Select columns that are relevant for the analysis

1. ID: "**UNITID**"
2. Institution name: "**INSTNM**"
3. City: "**CITY**"
4. State postcode: "**STABBR**"
5. Zip code: "**ZIP**"
6. School URL: "**INSTURL**"
7. Highest degree awarded: "**HIGHDEG**"
- 0: Non-degree granting
- 1: Certificate degree
- 2: Associates degree
- 3: Bachelors degree
- 4: Gradute degree
8. Control of institution: "**CONTROL**"
- 1: Public
- 2: Private non-proft
- 3: Private for-proft
9. Region: "**REGION**"
- 0: US service schools
- 1: New England (CT, ME, MA, NH, RI, VT)
- 2: Mid East (DE, DC, MD, NJ, NY, PA)
- 3: Great Lakes (IL, IN, MI, OH, WI)
- 4: Plains (IA, KS, MN, MO, NE, ND, SD)
- 5: Southeast (AL, AR, FL, GA, KY, LA, MS, NC, SC, TN, VA, WV)
- 6: Southwest (AZ, NM, OK, TX)
- 7: Rocky Mountains (CO, ID, MT, UT, WY)
- 8: Far west: (AK, CA, HI, NV, OR, WA)
- 9: Outlying areas (AS, FM, GU, MH, MP, PR, PW, VI)
10. Average cost of attendance (Academic year): "**COSTT4_A**"
11. In-state tuition & fees: "**TUITIONFEE_IN**"
12. Out-of-state tuition & fees: "**TUITIONFEE_IN**"
13. Average net price: "**NPT4_PUB**" && "**NPT4_PRIV**"
14. Net tuition revenue per FTE: "**TUITFTE**"
15. Instructional expenditure per FTE: "**INEXPFTE**"
16. Average faculty salary: "**AVGFACSAL**"

```python
# Select relevant columns

master_df = df[[
                "UNITID", "INSTNM", "CITY", "STABBR", "ZIP", "INSTURL", "HIGHDEG", "CONTROL", "REGION", "COSTT4_A", "TUITIONFEE_IN", "TUITIONFEE_OUT", "NPT4_PUB",
                "NPT4_PRIV", "TUITFTE", "INEXPFTE", "AVGFACSAL"
]]
master_df.head()
```
